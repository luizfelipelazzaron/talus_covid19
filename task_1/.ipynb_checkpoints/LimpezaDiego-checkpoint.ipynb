```python
#Imports
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('../dataset.xlsx')
```

```python
df
```

```python
#Dropando todas as colunas que só possuem valores nulos já que não adicionam ao projeto.
df = df.dropna(axis = 1, how = 'all')
df.head(10)
```

```python
#Dropando as linhas que não possuem valores ou então possuem poucos valores (Para esclarecimento, escolhi o minimo de valores 
#diferentes de nulo por linha).
for index, col in df.iterrows():
    counter = 0
    for row in col:
        if not row is np.nan:
            counter += 1
            if counter >= 5:
                #print(str(index) + " será mantido")
                break
        if counter < 5:
            df.drop(index)
            #print(str(index) + " será descartado")
            
df
```
